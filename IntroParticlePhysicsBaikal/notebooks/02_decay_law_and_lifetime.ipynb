# Decay law and lifetime

For an unstable state with lifetime $\tau$,

$$P(t)=e^{-t/\tau},\qquad \Gamma=\frac{1}{\tau}.$$

This notebook treats the lifetime as a fixed parameter and ignores production, boosts, detector acceptance, and backgrounds.

```python
from pathlib import Path
import sys

project_root = Path.cwd()
if not (project_root / "src").exists():
    project_root = project_root.parent
sys.path.insert(0, str(project_root / "src"))

import matplotlib.pyplot as plt
import numpy as np

from particle_intro.decay import survival_probability, decay_pdf
from particle_intro.plotting import plot_decay_law
```

```python
t = np.linspace(0.0, 8.0, 300)

fig, ax = plt.subplots(figsize=(7, 4))
plot_decay_law(ax, t=t, taus=[0.5, 1.0, 2.0])
ax.set_title("Exponential survival probability")
fig.tight_layout()
```

The decay-time density is normalized as

$$f(t)=\frac{1}{\tau}e^{-t/\tau},\qquad \int_0^\infty f(t)\,dt=1.$$

The decay width is a physical observable because it controls both the lifetime and, in scattering language, the resonance line shape.

```python
tau = 1.0
dt = t[1] - t[0]
finite_window_integral = np.sum(decay_pdf(t, tau)) * dt
finite_window_integral
```
