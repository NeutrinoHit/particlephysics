# Two-flavor neutrino oscillation probability

This notebook uses the vacuum two-flavor approximation:

$$P_{\nu_\mu\to\nu_\mu}=1-\sin^2 2\theta\,\sin^2\left(1.267\frac{\Delta m^2[\mathrm{eV}^2]L[\mathrm{km}]}{E[\mathrm{GeV}]}\right).$$

It is a pedagogical approximation, not a full atmospheric-neutrino analysis. Matter effects, three-flavor interference, detector resolution, and Earth geometry are not included.

```python
from pathlib import Path
import sys

project_root = Path.cwd()
if not (project_root / "src").exists():
    project_root = project_root.parent
sys.path.insert(0, str(project_root / "src"))

import matplotlib.pyplot as plt
import numpy as np

from particle_intro.oscillations import two_flavor_survival_probability
from particle_intro.plotting import plot_oscillation_probability
```

```python
L_over_E = np.logspace(1, 5, 500)
dm2 = 2.5e-3
theta = np.pi / 4

fig, ax = plt.subplots(figsize=(7, 4))
plot_oscillation_probability(ax, L_over_E=L_over_E, dm2_eV2=dm2, theta=theta)
ax.set_title("Atmospheric-scale two-flavor toy probability")
fig.tight_layout()
```

For atmospheric neutrinos, changing zenith angle changes the baseline from tens of km to roughly the Earth diameter. The relevant organizing variable is therefore $L/E$.

```python
examples = {
    "down-going, 1 GeV": (20.0, 1.0),
    "through Earth, 1 GeV": (12700.0, 1.0),
    "through Earth, 10 GeV": (12700.0, 10.0),
}

{label: two_flavor_survival_probability(L, E, dm2, theta) for label, (L, E) in examples.items()}
```
